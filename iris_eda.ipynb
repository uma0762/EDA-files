```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("iris.csv")
```

```python
sns.distplot(a= data['PetalWidthCm'], bins=20, color='m')
```

```python
sns.countplot(x='Species', data=data,palette='OrRd')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Species", y='SepalWidthCm', data=data, palette= 'YlGnBu')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Species", y='PetalWidthCm', data=data, palette= 'YlGnBu')
```

```python
plt.figure(figsize=(20,10))
sns.violinplot(x='Species', y ='PetalLengthCm', data=data, palette= 'OrRd')
```

```python
data.drop(['Id'], axis=1, inplace=True)
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data.corr(),cmap='YlGnBu', linecolor= 'white', annot = True,linewidths=1)
```

```python
sns.regplot(x='PetalWidthCm',y='PetalLengthCm', data=data)
```

```python
sns.regplot(x='SepalWidthCm',y='PetalLengthCm', data=data)
```

```python
axList= data.hist(bins=25, figsize=(8,8))

for ax in axList.flatten():
    if ax.is_last_row():
        ax.set_xlabel('size')
        
    if ax.is_first_col():
        ax.set_ylabel('Frequency')
```

```python
sns.set_context('talk')
sns.pairplot(data, hue='Species')
```

```python
sns.FacetGrid(data, hue="Species",size=6)\
    .map(sns.kdeplot, "PetalLengthCm")\
    .add_legend()
```
